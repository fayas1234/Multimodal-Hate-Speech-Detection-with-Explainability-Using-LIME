```python
import pandas as pd
import json
import os
import random
import clip
from PIL import Image
import numpy as np
import torch

# Load dataset from JSONL file
def load_dataset(file_path):
    dataset = []
    with open(file_path, 'r') as file:
        for line in file:
            example = json.loads(line)
            dataset.append(example)
    return dataset

# Define file paths
train_file = '/content/drive/My Drive/Hateful-memes/data/train.jsonl'
image_directory = '/content/drive/My Drive/Hateful-memes/data/'
```

```python
# Load dataset
train_dataset = load_dataset(train_file)

# Randomly select 20% of the dataset
subset_size = int(0.2 * len(train_dataset))
chosen_examples = random.sample(train_dataset, subset_size)

# Create DataFrame
data = []
for idx, example in enumerate(chosen_examples):
    text = example['text']
    img_path = os.path.join(image_directory, example['img'])
    label = example['label']
    data.append({'id': idx, 'img_text': text, 'img_path': img_path, 'label': label})

df = pd.DataFrame(data)

# Display DataFrame
print(df.head())
```

```python
# Load the CLIP model and processor
model, preprocess = clip.load("ViT-B/32")

# Define the function to extract features from images using CLIP
def extract_features(image_path):
    image = Image.open(image_path)
    image_input = preprocess(image).unsqueeze(0)
    with torch.no_grad():
        image_features = model.encode_image(image_input)
    image_features = image_features.cpu().numpy()
    return image_features

# Initialize variables for accuracy calculation
correct_predictions = 0
total_samples = len(df)
```

```python
# Iterate over the sampled DataFrame
for index, row in df.iterrows():
    image_path = row['img_path']
    label = row['label']

    # Extract features from the image using CLIP
    image_features = extract_features(image_path)

    # Convert the numpy array to a torch tensor
    image_features_tensor = torch.tensor(image_features).unsqueeze(0)  # Add a batch dimension

    # Forward pass through the model
    with torch.no_grad():
        logits_per_image, _ = model(image=image_features_tensor, text='')
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

    # Get the predicted label (0 or 1)
    predicted_label = np.argmax(probs)

    # Compare with the ground truth label
    if predicted_label == label:
        correct_predictions += 1

# Calculate accuracy
accuracy = correct_predictions / total_samples * 100
print(f"Accuracy: {accuracy:.2f}%")
```

```python
# Randomly select 20% of the dataset
subset_size = int(0.2 * len(train_dataset))
chosen_examples = random.sample(train_dataset, subset_size)

# Create DataFrame
data = []
for idx, example in enumerate(chosen_examples):
    text = example['text']
    img_path = os.path.join(image_directory, example['img'])
    label = example['label']
    data.append({'id': idx, 'img_text': text, 'img_path': img_path, 'label': label})

df = pd.DataFrame(data)

# Display DataFrame
print(df.head())

# Load the CLIP model and processor
model, preprocess = clip.load("ViT-B/32")

# Define the function to extract features from images using CLIP
def extract_features(image_path):
    image = Image.open(image_path)
    image_input = preprocess(image).unsqueeze(0)
    with torch.no_grad():
        image_features = model.encode_image(image_input)[0]  # Extracting only the image features
    return image_features

# Initialize variables for accuracy calculation
correct_predictions = 0
total_samples = len(df)

# Iterate over the sampled DataFrame
for index, row in df.iterrows():
    image_path = row['img_path']
    label = row['label']

    # Extract features from the image using CLIP
    image_features = extract_features(image_path)

    # Convert the numpy array to a torch tensor
    image_features_tensor = torch.tensor(image_features).unsqueeze(0)  # Add a batch dimension

    # Duplicate the single channel to have three channels
    image_features_tensor = image_features_tensor.expand(image_features_tensor.shape[0], 3, -1, -1)

    # Forward pass through the model
    with torch.no_grad():
        logits_per_image, _ = model(image=image_features_tensor, text='')
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

    # Get the predicted label (0 or 1)
    predicted_label = np.argmax(probs)

    # Compare with the ground truth label
    if predicted_label == label:
        correct_predictions += 1

# Calculate accuracy
accuracy = correct_predictions / total_samples * 100
print(f"Accuracy: {accuracy:.2f}%")
```
